```python
data_df = pd.read_csv('bikers_data.csv')
```

```python
data_y = data_df['Number of bikers'] # target
data_x = data_df.drop(['Number of bikers','Date'], axis=1) # input features
```

```python
# @title Default title text
# add bias =1
ones = np.ones(data_x.shape[0]) # to each row
# X with weights and bias
X = np.column_stack((ones, data_x.to_numpy()))

Y = data_y.to_numpy()
```

```python
# split
from sklearn.model_selection import train_test_split
# train = 80%
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2)
```

```python
# using sklearn model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
fit = model.fit(X_train, Y_train)
print(fit.score(X_train, Y_train))
```

```python
print(fit.score(X_test, Y_test))
```

```python
# train using normal equation
# Aw = y

w = np.linalg.inv(np.dot(X_train.T, X_train) )@ np.dot(X_train.T, Y_train)
w
# predict
prediction = X_test @ w
prediction

# eval the model
def mse(yhat, ytrue):
  error = ytrue - yhat
  square = error**2
  mse = square.mean()
  return mse

mse(prediction,Y_test)
```
